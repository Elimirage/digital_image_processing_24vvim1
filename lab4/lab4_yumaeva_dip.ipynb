```python
import sys
%matplotlib inline
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
image1 = cv.imread('/content/5Q29iqOV9ak.jpg')
gray_image1 = cv.cvtColor(image1, cv.COLOR_BGR2GRAY)
image = gray_image1
```

```python
threshold = 110
ret1, thresh1 = cv.threshold(image, threshold, 255, cv.THRESH_BINARY)
plt.figure(figsize=(70, 46))
plt.subplot(1, 3, 1)
plt.imshow(thresh1, 'gray')
plt.title('Global Thresholding (threshold = %d)' % threshold)
plt.xticks([])
plt.yticks([])
plt.show()
```
